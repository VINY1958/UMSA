 ## **<font color="#07a8ed">Codificación de variables categóricas</font>**

- job

- education
- default

```
analisis.head(3)
```

Se crea una función para reemplazar los nombres del mes por el número que corresponde.

```
def mes(month):
  if month == "jan":
    return int("1")
  elif month == "feb":
    return int("2")
  elif month == "mar":
    return int("3")
  elif month == "apr":
    return int("4")
  elif month == "may":
    return int("5")
  elif month == "jun":
    return int("6")
  elif month == "jul":
    return int("7")
  elif month == "aug":
    return int("8")
  elif month == "sep":
    return int("9")
  elif month == "oct":
    return int("10")
  elif month == "nov":
    return int("11")
  else:
    return int("12")
```

```
analisis.month = analisis.month.apply(mes)
```

```
imputacion = SimpleImputer(missing_values=3, strategy="mean")  ### la imputacion simple es por la media, la mediana, el valor mas frecuente, una constante u otro valor estadistico
```

### missing_value= El marcador de posición para los valores faltantes.
### strategy = La estrategia de imputación:
#### Si es “media”, entonces reemplace los valores faltantes utilizando la media a lo largo de cada columna.
###media-mediana-el valor mas frecuente, constante u otro valor estadistico

```
analisis.numeducation = imputacion.fit_transform(analisis[["numeducation"]])
```

```
analisis.numeducation = round(analisis.numeducation,3)
```

```
import pandas as pd
import numpy as np
```

```
import sklearn
from sklearn.preprocessing import LabelEncoder  ###Codifique las etiquetas de destino con valores entre 0 y n_classes-1
from sklearn.impute import SimpleImputer  ### imputacion simple
```

```
url = "https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/banco.csv"
```

```
analisis = pd.read_csv(url, index_col=0)
analisis.head(3)
```

```
analisis.job.value_counts()
```

```
analisis.info()
```

```
analisis.replace({"unknown": None}, inplace=True) ### reemplazo el unknown por el None
```

```
analisis.head(3)
```

```
analisis.job.value_counts().sum()
```

```
analisis.job.isnull().sum()
```

```
convertir = LabelEncoder()
```

```
analisis.job = convertir.fit_transform(analisis[["job"]])
```

```
analisis.job.value_counts()
```

```
analisis.job.value_counts().sum()
```

```
imputacion = SimpleImputer(missing_values=11, strategy="mean")
```

```
analisis.job = imputacion.fit_transform(analisis[["job"]])
```
