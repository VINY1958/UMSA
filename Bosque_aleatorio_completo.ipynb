```python
import pandas as pd
data = pd.read_csv("data.csv")
```

```python
data
```

```python
# Selecciono 60 filas aleatorias, valor semilla 0

datos_no_vistos = data.sample(n=60, random_state=0)
datos = data.drop(datos_no_vistos.index)
```

```python
datos_no_vistos.shape
```

```python
datos.shape
```

```python
# Regresion de PyCaret
from pycaret.regression import *
```

```python
# Setup, variable objetivo SalePrice, valor semilla 42
reg = setup(datos, target = 'SalePrice', session_id = 42)
```

```python
# Modelo Bosque aleatorio
arbol = create_model('rf')
```

```python
# Prediccion
prediccion = predict_model(arbol)
```

```python
prediccion
```

```python
# Guardar el modelo
save_model(arbol, 'arbol')
```

```python
# Leer el modelo
arbol_cargado = load_model('arbol')
```

```python
# Prediccion de datos no vistos
prediccion_no_vistos = predict_model(arbol_cargado, data = datos_no_vistos)
```

```python
prediccion_no_vistos.head()
```

```python
# Elimino SalePrice
datos_no_vistos = datos_no_vistos.drop('SalePrice', axis = 1)
```
